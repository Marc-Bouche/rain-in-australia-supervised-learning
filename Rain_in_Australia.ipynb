# **Grid Search - Hyperparameters**

### Step 1: Gather Model Performances Results

```python
# Function to calculate metrics from a confusion matrix
def evaluate_model(conf_matrix, y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    return accuracy, precision, recall, f1

# Calculate metrics for each model
metrics = {
    "KNN (no scaling)": evaluate_model(conf_matrix_KNN_ns, y_test, pred1),
    "KNN (scaling)": evaluate_model(conf_matrix_KNN_s, y_test, pred2),
    "KNN (k=3)": evaluate_model(conf_matrix_KNN_3, y_test, pred3),
    "Logistic Regression": evaluate_model(conf_matrix_lr, y_test, pred4),
    "Bagging": evaluate_model(conf_matrix_bagging, y_test, y_pred_bagging),
    "Pasting": evaluate_model(conf_matrix_pasting, y_test, y_pred_pasting),
    "Random Forest": evaluate_model(conf_matrix_rf, y_test, y_pred_rf),
    "Gradient Boosting": evaluate_model(conf_matrix_gb, y_test, y_pred_gb),
    "AdaBoost": evaluate_model(conf_matrix_ab, y_test, y_pred_ab),
}

# Display results
for model, (acc, prec, rec, f1) in metrics.items():
    print(f"{model}: Accuracy={acc:.2f}, Precision={prec:.2f}, Recall={rec:.2f}, F1 Score={f1:.2f}")
```

```python
# Convert the metrics dictionary to a DataFrame for easy plotting
metrics_df = pd.DataFrame(metrics, index=["Accuracy", "Precision", "Recall", "F1 Score"]).T

# Set the positions and width for the bars
x = np.arange(len(metrics_df.index))  # the label locations
width = 0.2  # width of the bars

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))

# Plot each metric as a separate set of bars
for i, metric in enumerate(metrics_df.columns):
    ax.bar(x + i * width, metrics_df[metric], width, label=metric)

# Add labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('Models')
ax.set_ylabel('Scores')
ax.set_title('Model Comparison by Evaluation Metrics')
ax.set_xticks(x + width * (len(metrics_df.columns) - 1) / 2)
ax.set_xticklabels(metrics_df.index, rotation=45)
ax.legend(title="Metrics")

# Display the plot
plt.tight_layout()
plt.show()
```

### Step 3: Apply Grid Search 

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100],  # Test a smaller and a standard number of trees
    'max_depth': [None, 10, 20],  # Limit depth to None, 10, and 20 to avoid overfitting
    'min_samples_split': [2, 5],  # Minimum samples required to split
    'min_samples_leaf': [1, 2]    # Minimum samples required at each leaf
}

# Initialize the Random Forest model
random_forest_model = RandomForestClassifier(random_state=0)

# Set up GridSearchCV
grid_search = GridSearchCV(
    estimator=random_forest_model,
    param_grid=param_grid,
    scoring='f1',  # scoring metric 
    cv=3,          # 3-fold cross-validation for quicker results
    n_jobs=-1      # Use all available CPU cores
)

# Fit the GridSearchCV
grid_search.fit(X_train_scaled, y_train)

# Get the best estimator
best_rf_model = grid_search.best_estimator_

# Print the best parameters &  best score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```
